# Advanced Exercise #5

```python
# Find top 3 correlated attributes to survive

df_num = df.select_dtypes(include = ['float64', 'int64']) 

df_num_corr = df_num.corr(method = 'kendall')['Survived'][2:]  # PassengerId & Survived not correlated & hence omitted
                                                             # Changing method to Kendall to get the required 3 correlate values
golden_features_list = df_num_corr[abs(df_num_corr) > 0.1].sort_values(ascending=False)
print("There are {} correlated values with Survived:\n{}".format(len(golden_features_list), golden_features_list))
```
